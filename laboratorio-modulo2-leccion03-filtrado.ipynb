# Laboratorio Filtrado de Datos

```python
# Carga el dataset de trabajo

import pandas as pd
```

```python
df_dataset = pd.read_csv("merged_netflix_data.csv")
df_dataset
```

```python
df_dataset.rename(columns= {"IMDB Score" :"IMDB_Score"} , inplace= True)
```

#### Ejercicio 1: ¿Cuáles son todos los títulos lanzados en 2021 que son del tipo 'Movie'?

```python
filtro1 = df_dataset["type"] == "Movie"
filtro2 = df_dataset["release_year"] == 2021

df_movie2021 = df_dataset[filtro1&filtro2]
df_movie2021

```

#### Ejercicio 2: ¿Qué títulos tienen una duración de 90 minutos? Muestra solo las columnas de título y de duración.

```python
filtro_duracion = df_dataset["duration"] == "90 min"
df_duracion_90 = df_dataset.loc[filtro_duracion, ["title","duration"]]
df_duracion_90
```

#### Ejercicio 3: ¿Cuáles son todos los títulos del país 'India' lanzados después de 2018?

```python
filtro_pais = df_dataset["country"] == "India"
filtro_anio = df_dataset ["release_year"] > 2018
filtro_combinado = filtro_pais & filtro_anio
titulos_india = df_dataset.loc[filtro_combinado,["title","release_year","country"]]
titulos_india
```

#### Ejercicio 4: ¿Qué títulos tienen clasificación 'TV-14' o 'R'?

```python
filtro_clasificacion = (df_dataset["rating"]== "TV-14") | (df_dataset["rating"] == "R")
titulos_clasificacion = df_dataset.loc[filtro_clasificacion,["title","rating"]]
titulos_clasificacion
```

#### Ejercicio 5: ¿Cuáles son los títulos dirigidos por 'Christopher Nolan' o 'Martin Scorsese'? ¿Cuántos a dirigido cada uno de los directores?

```python
filtro_directores = (df_dataset["director"]== "Christopher Nolan") | (df_dataset["director"]== "Martin Scorsese")
titulos_directores = df_dataset.loc[filtro_directores,["title","director"]]
conteo_directores = titulos_directores["director"].value_counts()

print(f"Titulos dirigidos por Christopher Nolan o Martin Scorsese: {titulos_directores}")
print(f"Numero de titulos dirigidos por cada director:{conteo_directores}")
```

#### Ejercicio 6: ¿Qué títulos pertenecen a los países 'United States', 'United Kingdom' o 'Canada'? Dentro de estos países, ¿cuántas películas tienen más de un lenguaje?

```python

filtro_paises = df_dataset["country"].isin(["United States", "United Kingdom", "Canada"])
titulos_paises = df_dataset.loc[filtro_paises, ["title", "country", "Language"]]
filtro_lenguajes = titulos_paises["Language"].str.contains(",")
conteo_multilenguaje = filtro_lenguajes.sum()

print(f"Títulos de los países seleccionados:{titulos_paises}")


print(f"Número de películas con más de un lenguaje: {conteo_multilenguaje}")
```

#### Ejercicio 7: ¿Qué títulos fueron lanzados entre 2015 y 2020, inclusive?

```python

filtro_anios = (df_dataset["release_year"] >= 2015) & (df_dataset["release_year"] <= 2020)
titulos_anios = df_dataset.loc[filtro_anios, ["title", "release_year"]]


print("Títulos lanzados entre 2015 y 2020:")
print(titulos_anios)
```

#### Ejercicio 8: ¿Cuáles son los títulos cuya puntuación en IMDb está entre 6.5 y 8.0 y que no sepamos cual es su duración?

```python

filtro_puntuacion = (df_dataset["IMDB_Score"] >= 6.5) & (df_dataset["IMDB_Score"] <= 8.0)
filtro_duracion = df_dataset["duration"].isna()


filtro_combinado = filtro_puntuacion & filtro_duracion

titulos_filtrados = df_dataset.loc[filtro_combinado, ["title", "IMDB_Score", "duration"]]


print("Títulos con puntuación entre 6.5 y 8.0 y duración desconocida:")
print(titulos_filtrados)
```

#### Ejercicio 9: ¿Cuáles son las películas cuya duración de 90 y 150 minutos?

```python
filtro_duracion = (df_dataset["duration"] == "90 min") | (df_dataset["duration"] == "150 min")

filtro_tipo = df_dataset["type"] == "Movie"

filtro_combinado = filtro_duracion & filtro_tipo


peliculas_duracion = df_dataset.loc[filtro_combinado, ["title", "duration", "type"]]


print("Películas con duración de 90 o 150 minutos:")
print(peliculas_duracion)
```

#### Ejercicio 10: ¿Cuáles son los títulos que contienen la palabra 'Love' en su título y que sean de tipo TV Show?

```python
filtro_titulo = df_dataset["title"].str.contains("Love", case=False, na=False)
filtro_tipo = df_dataset["type"] == "TV Show"
filtro_combinado = filtro_titulo & filtro_tipo

titulos_love_tvshow = df_dataset.loc[filtro_combinado, ["title", "type"]]


print("Títulos que contienen 'Love' y son de tipo TV Show:")
print(titulos_love_tvshow)
```

#### Ejercicio 11: ¿Cuáles son los títulos en los que el director tiene el apellido 'Johnson'?

```python

filtro_director = df_dataset["director"].str.contains(r"\bJohnson$", case=False, na=False)
titulos_directores_johnson = df_dataset.loc[filtro_director, ["title", "director"]]


print("Títulos dirigidos por alguien con apellido 'Johnson':")
print(titulos_directores_johnson)
```

#### Ejercicio 12: ¿Tenemos algún documental ('Documentaries') de Mike Tyson?

```python

filtro_documental = df_dataset["type"] == "Documentary"

filtro_mike_tyson = df_dataset["title"].str.contains("Mike Tyson", case=False, na=False) | df_dataset["director"].str.contains("Mike Tyson", case=False, na=False)

filtro_combinado = filtro_documental & filtro_mike_tyson

documentales_tyson = df_dataset.loc[filtro_combinado, ["title", "type", "director"]]


print("Documentales de Mike Tyson:")
print(documentales_tyson)
```

#### Ejercicio 13: Selecciona solo las columnas 'title', 'country' y 'release_year' del dataset.

```python

columnas_seleccionadas = df_dataset[["title", "country", "release_year"]]


print(f"Títulos, países y años de lanzamiento:{columnas_seleccionadas}")

```

#### Ejercicio 14: Filtra las columnas que contienen la palabra 'date'.

```python

columnas_date = df_dataset.filter(like="date", axis=1)


print(f"Columnas que contienen la palabra 'date':{columnas_date}")

```

#### Ejercicio 15: Usando `query()`, filtra todos los títulos del país 'France' con una duración de 80 minutos.

```python

titulos_francia_80min = df_dataset.query("country == 'France' and duration == '80 min'")


print("Títulos del país 'France' con duración de 80 minutos:")
print(titulos_francia_80min[["title", "country", "duration"]])
```

#### Ejercicio 16: Filtra los títulos que son películas ('Movie') y fueron lanzados antes de 2010.

```python

titulos_peliculas_antes_2010 = df_dataset.query("type == 'Movie' and release_year < 2010")


print("Películas lanzadas antes de 2010:")
print(titulos_peliculas_antes_2010[["title", "type", "release_year"]])
```

#### Ejercicio 17: Encuentra los 10 títulos con las puntuaciones más altas en IMDb.

```python
titulos_imdb = df_dataset[df_dataset["IMDB_Score"].notna()]


top_10_imdb = titulos_imdb.sort_values(by="IMDB_Score", ascending=False).head(10)


print("Top 10 títulos con las puntuaciones más altas en IMDb:")
print(top_10_imdb[["title", "IMDB_Score"]])
```

#### Ejercicio 18: Encuentra los 5 títulos con las puntuaciones más bajas en IMDb.

```python
titulos_imdb = df_dataset[df_dataset["IMDB_Score"].notna()]
bottom_5_imdb = titulos_imdb.sort_values(by="IMDB_Score", ascending=True).head(5)


print("Top 5 títulos con las puntuaciones más bajas en IMDb:")
print(bottom_5_imdb[["title", "IMDB_Score"]])
```

#### Ejercicio 19: Filtra todos los títulos que no tienen un director asignado (valores nulos en la columna 'director').

```python
titulos_sin_director = df_dataset[df_dataset["director"].isna()]


print("Títulos sin director asignado:")
print(titulos_sin_director[["title", "director"]])
```

#### Ejercicio 20: Filtra todos los títulos que tienen información completa en las columnas 'title', 'country' y 'IMDB Score'.

```python
titulos_completos = df_dataset.dropna(subset=["title", "country", "IMDB_Score"])


print("Títulos con información completa en 'title', 'country' y 'IMDB_Score':")
print(titulos_completos[["title", "country", "IMDB_Score"]])
```

#### Ejercicio 21: Identifica si hay títulos duplicados en la columna 'title'.

```python

titulos_duplicados = df_dataset[df_dataset.duplicated(subset=["title"], keep=False)]

# Mostrar los resultados
print("Títulos duplicados:")
print(titulos_duplicados[["title"]])
```
